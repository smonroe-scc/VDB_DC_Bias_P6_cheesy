```python
VJ=700.0E-3
VEE=-6.0
VEE_2=-4.0
VCC=8.0
```

Start with the following schematic:

The calculation for $V_{TH}$ is $$\frac{R_2 \cdot V_{CC} + V_{EE2} \cdot R_1}{R_1+R_2}$$

```python
R1=200.0E3
R2=36.0E3
VTH=(R2*VCC+R1*VEE_2)/(R1+R2)
print("VTH:",'{0:.5f}'.format(VTH),"V")
```

```python
VE=VTH-VJ
print("VE:",'{0:.5f}'.format(VE),"V")
```

```python
VC=-2.0
VRC=VCC-VC
IC=VRC/770.0
print("IC:",'{0:.5e}'.format(IC),"A")
```

```python
IE=IC
VRE=VE-VEE
RE=VRE/IE
print("RE :",'{0:.5f}'.format(RE),"\u03A9")
```
